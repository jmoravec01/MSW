```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
# METHODS (bisection & newton)
def bisection(f, a0, b0, eps=1E-10):
    a, b = a0, b0
    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x

def newton(f, a0, b0, delta=1E-10):
    a, b = a0, b0
    def df(x, h = 1E-10):
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew

def plot_result(f, a0, b0, x, title):
    print("root: ", x)
    plt.title(title)
    plt.plot(x, f(x), "bo")                                               # root
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")  # function
    plt.axhline(y=0)
    plt.show()
```

```python
# VARIABLES
func1 = lambda x: -x**3 + 3*x**2 + x -3
func2 = lambda x: 8*x - np.exp(-5*x) + 5
func3 = lambda x: np.sin(x+2)
```

```python
# FUNCTION 1
a1, b1 = -1.2, 3.2
plot_result(func1, a1, b1, bisection(func1, a1, b1), "Bisection: -x^3 + 3x^2 + x - 3")
plot_result(func1, a1, b1, newton(func1, a1, b1), "Newton: -x^3 + 3x^2 + x - 3")

timeb1 = timeit.timeit(lambda: bisection(func1, a1, b1), number=100000)
timen1 = timeit.timeit(lambda: newton(func1, a1, b1), number=100000)

print(f"TIME => BISECTION: {timeb1/100000}s, NEWTON: {timen1/100000}s")
```

```python
# FUNCTION 2
a2, b2 = -1, 1
plot_result(func2, a2, b2, bisection(func2, a2, b2), "Bisection: 8x - e^(-5x) + 5")
plot_result(func2, a2, b2, newton(func2, a2, b2), "Newton: 8x - e^(-5x) + 5")

timeb2 = timeit.timeit(lambda: bisection(func2, a2, b2), number=100000)
timen2 = timeit.timeit(lambda: newton(func2, a2, b2), number=100000)

print(f"TIME => BISECTION: {timeb2/100000}s, NEWTON: {timen2/100000}s")
```
