```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# BISECTION (interval halving method)
def bisection(fc, a, b, error_limit):

    def f(x):
        f = eval(fc)
        return f

    error = abs(b - a)

    while error > error_limit:
        c = (b+a)/2

        if f(a) * f(b) >= 0:
            print("Bisection will not work!")
            quit()
        
        elif f(c) * f(a) < 0:
            b = c
            error = abs(b-a)
        
        elif f(c) * f(b) < 0:
            a = c
            error = abs(b-a)
        
    
    print(f"The lower boundary is: {a} and the upper boundary is: {b}")

bisection("(3*x**2) + 3*x + 2", 0, 1, 0.5)

        
```

```python
def fxn1(x):
    return x**2 + 3*x -np.exp(x**(1/2)) 

xlist=np.linspace(0, 1, num=1000)
plt.figure(num=1, dpi=150)
plt.plot(xlist, fxn1(xlist), label = "Fxn1")
plt.plot(xlist, xlist*0, "--", label = "y=0")
plt.legend()

root = fsolve(fxn1, 0.5)
print(root)
```
